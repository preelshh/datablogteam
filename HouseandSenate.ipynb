```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("houseandsenate.csv")
#Used US History All House and Senate Members.csv file from Kaggle
data.head()
```

```python
term = data.congress
likelihood = data["nominate_log_likelihood"]
```

```python
fig = px.choropleth(data,
                     locations = "state_abbrev",
                     locationmode = "USA-states",
                     color = "nominate_log_likelihood",
                     color_continuous_scale = "ice",
                     scope = "usa",
                     animation_frame = "congress")
fig.update_layout(coloraxis_colorbar=dict(title="Likelihood of Being Nominated (log)"))
fig.update_layout(coloraxis_colorbar=dict(thickness=15, len=1))
fig.show()
```
